## D65

```python
d65 = np.loadtxt("./d65.illum")
x = np.linspace(300, 830, 531)
plt.plot(x, d65)
```

## RGB Camera Spectra

```python
cam_R, cam_G, cam_B = np.loadtxt("cam_R.spect"), np.loadtxt("cam_G.spect"), np.loadtxt("cam_B.spect")
x = np.linspace(380, 784, 101)
plt.plot(x, cam_R)
plt.plot(x, cam_G)
plt.plot(x, cam_B)
```

## CIE XYZ Spectral Sensitivity

```python
f = open('camera.kinect1.json', 'r')
xyz_dict = json.load(f)
XYZ_R = np.array(xyz_dict['sensitivities']['red'])[:, 1]
XYZ_G = np.array(xyz_dict['sensitivities']['green'])[:, 1]
XYZ_B = np.array(xyz_dict['sensitivities']['blue'])[:, 1]
x = np.linspace(365, 830, 94)
plt.plot(x, XYZ_R)
plt.plot(x, XYZ_G)
plt.plot(x, XYZ_B)
```
