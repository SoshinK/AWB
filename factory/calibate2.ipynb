```python
def pipeline2(raw, cct, T1, T3, T2, ccT1, ccT3, ccT2):
    if(cct >= ccT3):
        g = (1 / cct - 1 / ccT2) / (1 / ccT3 - 1 / ccT2)
        T = g * T3 + (1 - g) * T2
    else:
        g = (1 / cct - 1 / ccT3) / (1 / ccT1 - 1 / ccT3)
        T = g * T1 + (1 - g) * T3
    return transform(T, raw)    
```

```python
def pipeline(raw, cct, T1, T2, ccT1, ccT2):
    g = (1 / cct - 1 / ccT2) / (1 / ccT1 - 1 / ccT2)
    T = g * T1 + (1 - g) * T2
    return transform(T, raw)    
```

```python
mean_before = []
mean_after2CCT = []
mean_after3CCT = []

for i in tqdm_notebook(range(len(RAW))):
    img_raw = np.float64(np.asarray(mg.load(RAW[i])))
    img_gt = np.float64(np.asarray(mg.load(GT[i])))
    
    mask = np.loadtxt(MASK[i], delimiter=',')
    mask *= 2
    img_raw = wb(img_raw, mask)
    
    #img_3CCT = pipeline2(img_raw, CCT[i], T1, T3, T2, 3024.8659, 5019.4815, 5864.0833)
    img_2CCT = pipeline(img_raw, CCT[i], T1, T2, 3024.8659, 5864.0833)
    
    d_after2CCT = angdiff(img_2CCT, img_gt)
    #d_after3CCT = angdiff(img_3CCT, img_gt)
    d_before = angdiff(img_raw, img_gt)
    
    print("======\nT=", CCT[i], " (calibrate temps: ", 3024.8659, ", ", 5864.0833, ")")
    print("MAE before transform: ", d_before.mean(), " max error: ", np.amax(d_before))
    print("MAE after 2CCT transform: ", d_after2CCT.mean(), " max error: ", np.amax(d_after2CCT))
    #print("MAE after 3CCT transform: ", d_after3CCT.mean(), " max error: ", np.amax(d_after3CCT))
    
    mean_before.append(d_before.mean())
    mean_after2CCT.append(d_after2CCT.mean())
    #mean_after3CCT.append(d_after3CCT.mean())
    
print(mean_after2CCT)
print(mean_after3CCT)
print(mean_before)
```
